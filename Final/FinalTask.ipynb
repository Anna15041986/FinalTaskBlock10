Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

### Вывод 1:
Чем больше жилплощадь, тем выше  стоимость недвижимости.


```python
this_map = folium.Map(perfer_canvas=True)

def plotDot(point):
    folium.CircleMarker(
                        location=[point.lat, point.long],
                        radius=2).add_to(this_map)
df.apply(plotDot, axis=1)
this_map.fit_bounds(this_map.get_bounds())

this_map
```

Смотрим где расположена недвижимость.


```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

## Вывод 2:
Существует прямая зависимость цены недвижимости от количества комнат, жилой площади, количество ванных комнат, качества конструкции и дизайна.

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Distribution of price by floors');
```

## Вывод 3:
